```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data.csv')

print(veriSeti.head(30).to_string())

veriSeti = veriSeti.sort_values(by=['averageRating'], ascending=False).head(30)


filmler = veriSeti['title']
dereceler = veriSeti['averageRating']

plt.figure(figsize=(20, 7))
plt.bar(filmler, dereceler)
plt.xlabel('Filmler')
plt.ylabel('Dereceler')
plt.title('En Kt Dereceli 40 Film')
plt.xticks(rotation=90)
plt.ylim(1, 3)
plt.show()
```
